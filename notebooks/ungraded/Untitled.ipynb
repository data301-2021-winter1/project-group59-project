```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import collections
```

```python
df=pd.read_csv("chess_games.csv")
df.columns
```

```python
dfvicopen= df[['victory_status', 'opening_shortname']] 
dfvicopen
```

```python
dfvicopen.set_index([dfvicopen.index, 'victory_status'])['opening_shortname'].unstack()
```

```python
df1=dfvicopen.set_index([dfvicopen.index, 'victory_status'])['opening_shortname'].unstack()
df1[df1['Resign'].notna()]['Resign']
dfresign=df1[df1['Resign'].notna()]['Resign']
dfresign.to_frame()
resignframe=dfresign.to_frame()
collections.Counter(resignframe['Resign'])
resigndict=collections.Counter(resignframe['Resign'])
pd.DataFrame.from_dict(resigndict, orient='index')
resign=pd.DataFrame.from_dict(resigndict, orient='index')
```

```python
df1[df1['Out of Time'].notna()]['Out of Time']
dfoutoftime=df1[df1['Out of Time'].notna()]['Out of Time']
dfoutoftime.to_frame()
dfoutframe=dfoutoftime.to_frame()
collections.Counter(dfoutframe['Out of Time'])
outofdict=collections.Counter(dfoutframe['Out of Time'])
pd.DataFrame.from_dict(outofdict, orient='index')
outoftime=pd.DataFrame.from_dict(outofdict, orient='index')
```

```python
df1[df1['Mate'].notna()]['Mate']
dfmate=df1[df1['Mate'].notna()]['Mate']
dfmate.to_frame()
dfmateframe=dfmate.to_frame()
collections.Counter(dfmateframe['Mate'])
matedict=collections.Counter(dfmateframe['Mate'])
pd.DataFrame.from_dict(matedict, orient='index')
mate=pd.DataFrame.from_dict(matedict, orient='index')
```

```python
df1[df1['Draw'].notna()]
```

```python
dfdraw=df1[df1['Draw'].notna()]
dfdraw['Draw']
```

```python
dfdraw['Draw'].to_frame()
dfdrawframe=dfdraw['Draw'].to_frame()
dfdrawframe
```

```python
collections.Counter(dfdrawframe['Draw'])
drawdict=collections.Counter(dfdrawframe['Draw'])
drawdict
```

```python
pd.DataFrame.from_dict(drawdict, orient='index')
draw=pd.DataFrame.from_dict(drawdict, orient='index')
draw
```

```python
drawnew=draw.set_axis(['Draw'], axis=1)
resignnew=resign.set_axis(['Resign'], axis=1)
outoftimenew=outoftime.set_axis(['Out of Time'], axis=1)
matenew=mate.set_axis(['Checkmate'], axis=1)
draw.sort_index(ascending=True)
```

```python
resignnew.sort_index(ascending=True)
```

```python
pd.concat([drawnew, resignnew, matenew,outoftimenew] , ignore_index=True, axis=1)
stratoutcome=pd.concat([drawnew, resignnew, matenew,outoftimenew] , ignore_index=True, axis=1)
stratoutcome
```

```python
stratoutcome.sort_index(ascending=True)
```

```python
stratoutcome.set_axis(['Draw', 'Resign','Checkmate', 'Out of Time'], axis=1)
stratoutcomenew=stratoutcome.set_axis(['Draw', 'Resign','Checkmate', 'Out of Time'], axis=1)
stratoutcomenew.fillna(0)
beforestats=stratoutcomenew.fillna(0)
beforestats
```

```python
beforestats["Total Games"]=beforestats.sum(axis=1)
beforestats
```

```python
column_list=list(beforestats)
dfsum=beforestats[column_list].sum(axis=1)
```

```python
dfsum
```

```python
pd.DataFrame(dfsum, columns= ['Total Games'])
totalgames=pd.DataFrame(dfsum, columns= ['Total Games'])
```

```python
pd.concat([beforestats, totalgames], ignore_index=True, axis=1)
new=pd.concat([beforestats, totalgames], ignore_index=True, axis=1)
new
```

```python
new.set_axis(['Draw', 'Resign','Checkmate', 'Out of Time', 'Total Games'], axis=1)
NEWDATA=new.set_axis(['Draw', 'Resign','Checkmate', 'Out of Time', 'Total Games'], axis=1)
```

```python
NEWDATA.to_csv('New_Data.csv')
```
